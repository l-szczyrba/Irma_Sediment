## Import libraries

```python
import pandas as pd
import numpy as np
import pyexcel_xlsx
```

## Read excel and create dataframe

```python
data = pd.read_excel(r'/Users/lszczyrba/Project_3.6.3/HelloWorld/Irma_Sediment/IrmaMudThicknessComparisonsJuneUpdate.xlsx')
df = pd.DataFrame(data)
print(df.head())
                     
```

## Drop rows with a null value 

```python
no_null = df.dropna()
#print(no_null)
df = no_null
print(df)
```

## Drop rows where 'ThicknessD1cm' = 0

```python
df = df[df['ThicknessD1cm'] != 0]
print(df)

    
```

## Drop rows with column 12 ('Notes') = 1

```python
print(df[df.Notes == 1])
```

## Delete 'Notes' column (because it is useless now)

```python
df = df.drop(['Notes'], axis=1)
print(df)
CleanData = df
```

## Export a copy of cleaned data to excel

```python
# Create a Pandas Excel writer
writer = pd.ExcelWriter('CleanData.xlsx')
# Convert the dataframe to an XlsxWriter Excel object.
df.to_excel(writer, sheet_name='Sheet1')
# Close the Pandas Excel writer and output the Excel file
writer.save()
```

## Separate and export data into training and evaluation data

```python
print(len(df))
print(int(len(df)) * 0.8)
df_training = df[0:63]
print(df_training)
df_evaluation = df[63:len(df)]
print(df_evaluation)


writer1 = pd.ExcelWriter('Training.xlsx')
df_training.to_excel(writer1, sheet_name='Sheet1')
writer1.save()

writer2 = pd.ExcelWriter('Evaluation.xlsx')
df_evaluation.to_excel(writer2, sheet_name='Sheet1')
writer2.save()
```
