```python
import pandas as pd
import numpy as np
import pyexcel_xlsx
import matplotlib.pyplot as plt
%pylab inline
```

```python
data = pd.read_excel(r'/Users/lszczyrba/Project_3.6.3/HelloWorld/Irma_Sediment/Distances.xlsx', header = [0])
df = pd.DataFrame(data)
```

```python
df.insert(0,'ID', range(0,len(df)))
df.set_index('ID')
no_null = df.dropna()
df = no_null
```

```python
df = df[df['ThickD1cm'] != 0]
df.info()
df
```

```python
df['ThickD1cm'].replace(regex= True, inplace=True,to_replace=r'<',value=r'')
df['ThickD2cm'].replace(regex= True, inplace=True,to_replace=r'<',value=r'')
df['ThickD3cm'].replace(regex= True, inplace=True,to_replace=r'<',value=r'')
```

```python
df["ThickD1cm"] = df.ThickD1cm.astype(float)
df["ThickD2cm"] = df.ThickD2cm.astype(float)
df["ThickD3cm"] = df.ThickD3cm.astype(float)
df['ThickD3cm'].replace(inplace=True,to_replace=20,value=1)
df.info()
df
```

```python
grouped = df.groupby('Site')
grouped.size()
```

```python
BW2 = df[df['Site'] == 'BW2']
BW3 = df[df['Site'] == 'BW3']
CBD = df[df['Site'] == 'CBD']
CCB = df[df['Site'] == 'CCB']
SNK = df[df['Site'] == 'SNK']
UFU = df[df['Site'] == 'UFU']
WTZ = df[df['Site'] == 'WTZ']
```

```python
BW2 = BW2.sort_values(by = ['Distance'])
BW3 = BW3.sort_values(by = ['Distance'])
CBD = CBD.sort_values(by = ['Distance'])
CCB = CCB.sort_values(by = ['Distance'])
SNK = SNK.sort_values(by = ['Distance'])
UFU = UFU.sort_values(by = ['Distance'])
WTZ = WTZ.sort_values(by = ['Distance'])
```

```python
BW2_fig = BW2.plot.line(x = 'Distance', y = ['ThickD1cm', 'ThickD2cm', 'ThickD3cm'], marker = 'o', figsize = (15,15), title = 'BW2')
plt.savefig('BW2_Distance_Dates.png')

BW3_fig = BW3.plot.line(x = 'Distance', y = ['ThickD1cm', 'ThickD2cm', 'ThickD3cm'], marker = 'o', figsize = (15,15), title = 'BW3')
plt.savefig('BW3_Distance_Dates.png')

CBD_fig = CBD.plot.line(x = 'Distance', y = ['ThickD1cm', 'ThickD2cm', 'ThickD3cm'], marker = 'o', figsize = (15,15), title = 'CBD')
plt.savefig('CBD_Distance_Dates.png')

CCB_fig = CCB.plot.line(x = 'Distance', y = ['ThickD1cm', 'ThickD2cm', 'ThickD3cm'], marker = 'o', figsize = (15,15), title = 'CCB')
plt.savefig('CCB_Distance_Dates.png')

SNK_fig = SNK.plot.line(x = 'Distance', y = ['ThickD1cm', 'ThickD2cm', 'ThickD3cm'], marker = 'o', figsize = (15,15), title = 'SNK')
plt.savefig('SNK_Distance_Dates.png')

UFU_fig = UFU.plot.line(x = 'Distance', y = ['ThickD1cm', 'ThickD2cm', 'ThickD3cm'], marker = 'o', figsize = (15,15), title = 'UFU')
plt.savefig('UFU_Distance_Dates.png')

WTZ_fig = WTZ.plot.line(x = 'Distance', y = ['ThickD1cm', 'ThickD2cm', 'ThickD3cm'], marker = 'o', figsize = (15,15), title = 'WTZ')
plt.savefig('WTZ_Distance_Dates.png')
```
